```python
driver = webdriver.Chrome()
driver.get("https://www.bls.gov/charts/consumer-price-index/consumer-price-index-average-price-data.htm")
#driver.maximize_window()
print(driver.title)
```

```python
button = driver.find_element(By.XPATH, '//*[@id="bodytext"]/div[3]/div[2]/a') 
button.click()
```

```python
Month = []  #/p
Chicken =[] #5
Eggs =[]    #7
Ground_Beef =[]  # 9
Milk =[]   # 11

drive = driver.find_elements(By.TAG_NAME,'tr')


for i in range(1, len(drive)):
    try:
        chicken_cell = driver.find_element(By.XPATH, f'//*[@id="cpi_rc_5cpiavprd"]/tbody/tr[{i}]/td[5]')
        Chicken.append(chicken_cell.text)

        eggs_cell = driver.find_element(By.XPATH, f'//*[@id="cpi_rc_5cpiavprd"]/tbody/tr[{i}]/td[7]') 
        Eggs.append(eggs_cell.text)

        ground_beef_cell = driver.find_element(By.XPATH, f'//*[@id="cpi_rc_5cpiavprd"]/tbody/tr[{i}]/td[9]') 
        Ground_Beef.append(ground_beef_cell.text)

        milk_cell = driver.find_element(By.XPATH, f'//*[@id="cpi_rc_5cpiavprd"]/tbody/tr[{i}]/td[11]')  
        Milk.append(milk_cell.text)

        month_cell = driver.find_element(By.XPATH, f'//*[@id="cpi_rc_5cpiavprd-{i}-0"]/p') 
        Month.append(month_cell.text) 


    except:
        print(f"Error extracting data from row {i}") 

```

```python
data = {
    'month': Month,
    'chicken': Chicken,
    'eggs': Eggs,
    'ground_beef': Ground_Beef,
    'milk': Milk
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
month = []
year = []

for index, row in df.iterrows():  # Use iterrows() to loop through rows
    date_str = row['month']  
    month_str, year_str = date_str.split()  # Split the string by space
    month.append(month_str)
    year.append(year_str)


df['month'] = month
df['year'] = year
```
